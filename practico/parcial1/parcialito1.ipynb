b. Queremos calcular $p(N_b(1) \ge 3 | N(1) = 7) = 1 - p(N_b(1) \lt 3 | N(1) = 7)$

Ahora, debemos tener en cuenta Bayes:

$ p(N_b(1) \lt 3| N(1) = 7) = \frac{p(N_b(1) < 3, N(1) = 7)}{p(N(1) = 7)} $

Pero $N_b$ y $N$ no son independientes, entonces tengo que hacerlos disjuntos de alguna forma:

$ p(N_b(1) < 3, N(1) = 7) = p(N_b(1) < 3, N_b(1) + N_c(1) = 7) = \sum_{k=0}^{2} p(N_b(1) = k, N_c(1) = 7-k) = \sum_{k=0}^{2} p(N_b(1) = k)p(N_c(1) = 7-k) = \sum_{k=0}^{2} \frac{e^{-2}2^k}{k!}\frac{e^{-3}3^{(7-k)}}{(7-k)!})$

Y reemplazamos arriba para poder calcular

```python
suma = 0
for k in range(3):
    suma += exp(-2) * 2**k / factorial(k) * exp(-3) * 3**(7-k) / factorial(7-k)

1 - suma / (exp(-5) * 5**7 / factorial(7))
```
