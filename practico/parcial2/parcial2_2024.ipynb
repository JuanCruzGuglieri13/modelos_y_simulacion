```python
# b)
p = [0.13, 0.22, 0.35, 0.3]
def algo_x(p):
    while True:
        y = int(random() * 4)
        u = random()
        if u < p[y] / 0.35:
            return y
        


import matplotlib.pyplot as plt
from collections import Counter

def plot_frecuencias(muestras, label):
    conteos = Counter(muestras)
    xs = sorted(conteos.keys())
    ys = [conteos[x] / len(muestras) for x in xs]
    plt.bar(xs, ys, alpha=0.6, label=label)

# Parámetro de simulación
n_sim = 10000
# Simulación con cada método
muestras = [algo_x(p) for _ in range(n_sim)]
# Plot
plt.figure(figsize=(10, 6))
plot_frecuencias(muestras, "Aceptación y Rechazo")

plt.xticks(range(5))
plt.xlabel("Valor simulado de X")
plt.ylabel("Frecuencia relativa")
plt.title("Frecuencia relativa de X")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def TIX():
    u = random()
    if u < 2/3:
        return (1.5 * u) ** (2/3)
    else:
        return 3*u - 1
```

```python
iteraciones = [TIX() for _ in range(100000)]
# Plotar histograma
plt.hist(iteraciones, bins=100, density=True, alpha=0.6, color='skyblue', edgecolor='black')
plt.title('Histograma de 100.000 muestras - Transformada Inversa')
plt.xlabel('Valor generado')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()

# Función de densidad teórica
def f_teorica(x):
    return np.piecewise(x, 
                        [ (0 <= x) & (x < 1), 
                          (1 <= x) & (x <= 2)],
                        [lambda x: x**(1/2), 
                         lambda x: 1/3,
                         0])

# Dominio extendido para visualizar toda la densidad
x_vals = np.linspace(0, 2, 1000)
y_vals = f_teorica(x_vals)

# Graficar la densidad teórica
plt.plot(x_vals, y_vals, label='f(x) teórica', color='darkblue')
plt.title('Función de Densidad Teórica')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
# b)
Nsim = 10000
count = 0
for _ in range(Nsim):
    x = TIX()
    if x > 4:
        count += 1
print(f"Frecuencia de x > 4: {count / Nsim:.4f}")
```

```python
def lambda_t(t):
    if 0 <= t < 3:
        return 5 + 5*t
    elif 3 <= t <= 5:
        return 20
    elif 5 < t <= 9:
        return 30 - 2 * t


def hot_dog(T):
    Eventos = []
    intervalos = [(0, 1), (1, 2), (2, 6), (6, 8), (8, 9)]

    for (t_ini, t_fin) in intervalos:
        ts = np.linspace(t_ini, t_fin, 100)
        lambda_max_local = max(lambda_t(t) for t in ts)
        print(lambda_max_local)
        u = 1 - random()
        t = -(np.log(u) / lambda_max_local) + t_ini
        while t <= t_fin:
            v = random()
            if v < lambda_t(t) / lambda_max_local:
                Eventos.append(t)
            t += -np.log(1 - random()) / lambda_max_local

    return Eventos

def estimar_promedio(N_sim):
    return np.mean([np.mean(hot_dog(9)) for _ in range(N_sim)])

# Simulación
print("Estimación del promedio de tiempo entre eventos:")
N_sim = 1000
promedio_estimado = estimar_promedio(N_sim)
print(f"Promedio estimado: {promedio_estimado:.4f} minutos")
```
